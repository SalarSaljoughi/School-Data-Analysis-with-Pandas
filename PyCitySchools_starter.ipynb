```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#unique_school_names = school_data_complete["school_name"].value_counts()
#unique_school_names


school = school_data_complete['school_name'].nunique()
school

student_data = school_data_complete['student_name'].count()
student_data


total_budget = school_data['budget'].sum()
total_budget

average_math = school_data_complete['math_score'].mean()
average_math


average_reading = school_data_complete['reading_score'].mean()
average_reading

overal_passing = (average_math + average_reading) / 2
overal_passing


df_one = school_data_complete[school_data_complete.math_score >= 70]
percent_passing_math = (df_one['math_score'].count() / (student_data))*100
percent_passing_math

df_two = school_data_complete[school_data_complete.reading_score >= 70]
percent_passing_reading = (df_two['reading_score'].count() / (student_data))*100
percent_passing_reading


df = pd.DataFrame ( [[school, student_data , total_budget , average_math , average_reading , percent_passing_math , percent_passing_reading , overal_passing]] , columns = ["Total School", "Total Students", "Total Budget", "Average Math Score" , 
                              "Average Reading Score" , "% Passing Math" , "% Passing Reading" , "Overal Passing"])

df
```
